Above code was to clean up the noisyTextDescription data. Cleaned up data had a lower accuracy compared to the original noisyTextDescription data so I used the original noisyTextDescription data instead

```python
from keras.preprocessing.text import Tokenizer
from keras.preprocessing.sequence import pad_sequences 
# The maximum number of words to be used. (most frequent)
MAX_NB_WORDS = 50000
# Max number of words in each complaint.
MAX_SEQUENCE_LENGTH = 250
# This is fixed.
EMBEDDING_DIM = 100
tokenizer = Tokenizer(num_words=MAX_NB_WORDS, filters='!"#$%&()*+,-./:;<=>?@[\]^_`{|}~', lower=True)
tokenizer.fit_on_texts(df['noisyTextDescription'].values)
word_index = tokenizer.word_index
print('Found %s unique tokens.' % len(word_index))
```

Tokenizing and Padding the train text (noisyTextDescription)

```python
X = tokenizer.texts_to_sequences(df['noisyTextDescription'].values)
X = pad_sequences(X, maxlen=MAX_SEQUENCE_LENGTH)
print('Shape of data tensor:', X.shape)
```

One hot encoding the categories

```python
Y = pd.get_dummies(df['category']).values
print('Shape of label tensor:', Y.shape)
```

Assigning the training text data and training labels to X_train and Y_train respectively

```python
X_train, Y_train = X, Y
print(X_train.shape,Y_train.shape)
```

Tokenizing and Padding the test text (noisyTextDescription)

```python
X_test = tokenizer.texts_to_sequences(df_test['noisyTextDescription'].values)
X_test = pad_sequences(X_test, maxlen=MAX_SEQUENCE_LENGTH)
print('Shape of data tensor:', X_test.shape)
```

LSTM Model to for classification using text data

```python
from keras.models import Sequential
from keras.layers.embeddings import Embedding
from keras.layers import LSTM, Dense, Dropout, SpatialDropout1D
from keras.callbacks import ModelCheckpoint, EarlyStopping

model1 = Sequential()
model1.add(Embedding(MAX_NB_WORDS, EMBEDDING_DIM, input_length=X.shape[1]))
model1.add(SpatialDropout1D(0.2))
model1.add(LSTM(100, dropout=0.2, recurrent_dropout=0.2))
model1.add(Dense(27, activation='softmax'))
model1.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

epochs = 5
batch_size = 64
print(X_train.shape,Y_train.shape)
history = model1.fit(X_train, Y_train, epochs=epochs, batch_size=batch_size,validation_split=0.1,callbacks=[EarlyStopping(monitor='val_loss', patience=3, min_delta=0.0001)])
```

Save predictions of the test set from LSTM model in pred_text

```python
pred_text = model1.predict(X_test)
```

```python
import cv2
import os

def image_loading(folder):
    images = []
    file = []
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename))
       # print("fileanme",filename)
        if img is not None:
            images.append(img/255.0)
            file.append(filename)
    return images, file
folder="/kaggle/input/uw-cs480-fall20/suffled-images/shuffled-images/"
```

```python
# example of using a pre-trained model as a classifier
from keras.preprocessing.image import load_img
from keras.preprocessing.image import img_to_array
from keras.applications.vgg16 import preprocess_input
from keras.applications.vgg16 import decode_predictions
from keras.applications.vgg16 import VGG16
from keras.layers import Dense, Flatten
from keras.models import Model
```

Using Pre-trained VGG16 model for image classification

```python
model = VGG16(include_top=False, input_shape=(80, 60, 3))
flat1 = Flatten()(model.layers[-1].output)
class1 = Dense(1024, activation='relu')(flat1)
output = Dense(27, activation='softmax')(class1)
model = Model(inputs=model.inputs, outputs=output)
model.get_layer('block1_conv1').trainable = False
model.get_layer('block1_conv2').trainable = False
model.get_layer('block2_conv1').trainable = False
model.get_layer('block2_conv2').trainable = False
```

Freezing first 4 layers

```python
for layer in model.layers[:4]:
    layer.trainable = False
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```
